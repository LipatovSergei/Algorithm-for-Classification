##Let's use OneHotEncoder

```python
df.shape
```

```python
ohe = OneHotEncoder(sparse=False)
columns_to_encode = ['region', 'fuel', 'manufacturer', 'transmission', 'short_model', 'age_category', 'state']

for column in columns_to_encode:
    ohe = OneHotEncoder(sparse=False)
    ohe.fit(df[[column]])
    ohe_features = ohe.transform(df[[column]])
    feature_names = ohe.get_feature_names_out()
    df[feature_names] = ohe_features

    # Удаление исходной колонки
    df.drop(columns=[column], inplace=True)
```

#Modelling

```python
df_prepared = df.copy()
df_prepared = df_prepared.drop(['price_k$', 'posting_date'], axis=1)

x = df_prepared.drop(['price_category'], axis=1)
y = df_prepared['price_category']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
```

##Random Forest Classifier

```python
%%timeit
rf_clf = RandomForestClassifier()
rf_clf.fit(x_train, y_train)
predicted_train_rf = rf_clf.predict(x_train)
predicted_test_rf = rf_clf.predict(x_test)
print(accuracy_score(y_test, predicted_test_rf))
```

```python
f_imp_list = list(zip(x_train.columns, rf_clf.feature_importances_))
f_imp_list.sort(key = lambda x: x[1], reverse=True)
f_imp_list[:7]
```

##LogisticRegression

```python
logreg = LogisticRegression(penalty= 'l2', solver = 'newton-cg', max_iter=250)
logreg.fit(x_train, y_train)
logreg.score(x_train, y_train)
```

```python
logreg_pred = logreg.predict(x_test)
accuracy_score(y_test, logreg_pred)
```

```python
confusion_matrix(y_test, logreg_pred)
```

```python
cross_validate_logreg = cross_validate(logreg, x, y, cv=5, scoring='precision')
```

```python
cv_lr_score = cross_val_score(logreg, x, y, cv=5)
```

##Multi-layer Perceptron classifier

```python
%%timeit
mlp = MLPClassifier(random_state=42, alpha=0.01, hidden_layer_sizes=200, activation = 'tanh')
mlp.fit(x_train, y_train)
mlp.score(x_train, y_train)
pred_mlp = mlp.predict(x_test)
print(accuracy_score(y_test, pred_mlp))
```

##Let's do cross validation for all three models

```python
for model in [rf_clf, logreg, mlp]:
  cv_score = cross_val_score(model, x, y, cv=5)
  print(cv_score.mean(), cv_score.std())
```
